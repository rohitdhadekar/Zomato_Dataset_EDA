```python
data = pd.read_excel('/content/Country-Code.xlsx')
```

```python
data.head()
```

```python
final_df = pd.merge(df, data, on ='Country Code', how = 'left' )
final_df.head()
```

```python
country_names = final_df.Country.value_counts().index
country_values = final_df.Country.value_counts().values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct= '%1.2f%%');
plt.axis('square');
```

**Observation**:  Major transactions are from India followed by United States and United Kingdom.



```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns= {0:'Rating Count'})
```

```python
ratings
```

## Observation:
1. When the ratings are in between 4.5 to 4.9 ---> Excellent
2. When the ratings are in between 4.0 to 4.4 ---> Very Good
3. When the ratings are in between 3.5 to 3.9 ---> Good
4. When the ratings are in between 2.5 to 3.4 ---> Average
5. When the ratings are in between 0.1 to 2.4 ---> Poor
6. 0 rating is given by many of the people.

```python
ratings.columns
```

```python
plt.figure(figsize=(10,10))
sb.barplot(data= ratings, x='Aggregate rating', y= 'Rating Count')
```

```python
plt.figure(figsize=(10,10))
sb.barplot(data= ratings, x='Aggregate rating', y= 'Rating Count',hue ='Rating color',palette= ['White','Red','Orange','Yellow','Green','Green'] );
```

**Observation:**
1. Not rated count is very high.
2. Maximum number of ratings are in between 2.8 to 3.9.



```python
ratings.columns
```

```python
sb.countplot(data= ratings , x= 'Rating color',palette= ['White','Red','Orange','Yellow','Green','Green'] );
```
